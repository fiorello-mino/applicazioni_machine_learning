```python
import pandas as pd
import numpy as np
```

Esercizio 1: leggo il file excel in una matrice numpy

```python
patterns_excel = pd.read_excel('PCA-Data-Esercise.xlsx') #reading the file
matrix = patterns_excel.to_numpy() #converting to numpy matrix

X = matrix[2:, 3:]
X
```

2. Calcolo la matrice di correlazione del set di dati

```python
corr_matrix = np.corrcoef(X, rowvar=False) #computing correlation matrix
corr_matrix
```

3. Diagonalizzo la matrice di correlazione

```python
eigenvalues, eigenvectors = np.linalg.eig(corr_matrix) #computing eigenvalues and eigenvectors

D = np.diag(eigenvalues) #diagonal matrix
Q = eigenvectors #orthonormal matrix
print("Autovalori: ", eigenvalues)
print("Autovettori: ", eigenvectors)
print(Q)
```

4. Ordino in modo decrescente gli autovalori e gli autovettori corrispondenti

```python
idx = np.argsort(eigenvalues)[::-1]
eigenvalues_desc = eigenvalues[idx]
eigenvectors_desc = eigenvectors[:,idx]
D_desc = np.diag(eigenvalues_desc)

print("D decrescente:\n", D_desc)
print("Q decrescente:\n", eigenvectors_desc)

```

5. Scelgo quanti autovalori considerare nella PCA

```python
total_var = np.sum(eigenvalues_desc)
cum_perc = np.cumsum(eigenvalues_desc) / total_var

print("Peso cumulativo (%):", cum_perc)
```

6. PCA: moltiplico i dati iniziali per gli autovettori degli autovalori scelti

```python
k = 2
P = eigenvectors_desc[:, :k]

X_reduced = np.dot(X, P)
X_reduced
```
